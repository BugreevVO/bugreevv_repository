Датафрейм состоит из 22 столбцов и содержит информацию о 23699 объявлениях о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет.

При первичном знакомстве с данными обнаружены следующие проблемы:
- название столбца "расстояние до центра города (м)" не соответствует правилу "хорошего стиля";
- пропущенные значения присутствуют в 14 из 22 столбцов;
- некорректные типы данных в 5 из 22 столбцов.

На этапе предобработки, для корректного анализа, нам предстоит заполнить все пропущенные значения, изменить типы данных.

## Предобработка данных

### Стиль заголовков

```python
# заменим название столбца 'cityCenters_nearest' применив метод .rename()
data = data.rename(columns={'cityCenters_nearest': 'city_centers_nearest'})
```

Следуя правилам "хорошего стиля" изменили название столбца "расстояние до центра города".

### Обработка пропусков

```python
# зададим цикл, который пройдет по столбцам,
# в которых пустые значения заменим на медианы
for i in (
    ['ceiling_height', 'airports_nearest', 'city_centers_nearest',
    'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest',
    'floors_total']):
    data.loc[data[i].isna(), i] = data[i].median()
```

```python
# заменим все значения, где значение в столбце "этаж" превысило значение в
# столбце "всего этажей в доме", прировняем его колличеству этажей в доме
data.loc[data['floors_total'] < data['floor'], 'floors_total'] = data.loc[data['floors_total'] < data['floor'], 'floor']
```

```python
# применим метод .fillna() для замены пустых значений в столбцах "жилая площадь" и "площадь кухни" на
# медиану с разбивкой по количеству комнат, с помощью группировки методом .groupby()
data['living_area'] = data['living_area'].fillna(data.groupby('rooms')['living_area'].transform('median'))
```

```python
# заменим пропущеные значения в столбце 'kitchen_area' на "0" для студий и квартир свободной планировки
data.loc[(data['studio'] == True) |
       (data['open_plan'] == True) &
       (data['kitchen_area'].isna()), 'kitchen_area'] = 0
# оставшиеся пропуски заменим медианой
data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data['kitchen_area'].median()
```

```python
# применим метод .fillna() для замены пропущеных значений в столбце "апартаменты" на False
data['is_apartment'] = data['is_apartment'].fillna(False)
```

Пропущенные значения в столбце "апартаменты" заменим булевым значением False, т.к. недвижимость со статусом "апартаменты" довольно нечастое явление и при подаче объявления о продаже, продавец должен указать данную особенность. Соответственно мы сделали вывод, что значение пропущено в связи с тем, что продавец пропустил данную графу в обьявлении намеренно, т.к. она никак не относится к его недвижимости.

```python
# применим метод .fillna() для замены пропущеных значений в столбце "апартаменты" на "0"
data['balcony'] = data['balcony'].fillna('0')
```

Пропуски в столбце "число балконов" это, скорее всего их отсутствие (0 балконов). Люди могут не знать сколько у них этажей в доме или точную высоту потолков, но количество банконов они знают точно, т.к. его наличие сильно влияет на цену. Поэтому, если они его не указывают, то, вероятней всего, балконов нет.

```python
# оставим только те строки, где значение "название населённого пункта" не пустое
data = data[(~data['locality_name'].isna())]
```

Количество строк в с пропуском в столбце "название населённого пункта" удалили, это не окажет существенное влияние на результаты анализа, т.к. их мало (49 объявлений).

```python
# построим график зависимости числа пропусков в столбце "сколько дней было размещено объявление" от даты публикации объявления.
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'])
data.query('days_exposition.isna()')['first_day_exposition'].hist(bins = 20, figsize = (10,5))
plt.show()
```

Число пропусков растет к концу наблюдаемого периода. Вероятно, пропуски означают незакрытые объявления.
Будем считать, что дата выгрузки данных равна последнему дню наблюдений.

```python
# создадим переменную, которую будем использовать в расчете значений для заполнения пропусков в столбце days_exposition
max_dt = data['first_day_exposition'].max()

# заполняем пропуски разностью first_day_exposition и max_dt в днях
data.loc[data['days_exposition'].isna(),
        'days_exposition'] = (max_dt - data.loc[data['days_exposition'].isna(), 'first_day_exposition']).dt.days
```

```python
data.isna().sum()
```

Мы избавились от всех пропусков в датафрейме.

Теперь избавимся от неявных дупликатов в столбце с названиями населенных пунктов.

```python
# создадим словарь для замены значений в названиях населённых пунктов
text_to_replace = {'ё': 'е',
                   'деревня': '', 
                   'коттеджный поселок': '',
                   'городской поселок': '',
                   'поселок городского типа': '',
                   'посёлок станции': '',
                   'поселок станции': '',
                   'посёлок при железнодорожной станции': '',
                   'поселок при железнодорожной станции': '',
                   'поселок': '',
                   'садовое товарищество': '',
                   'садоводческое некоммерческое товарищество': '',
                   'село': '',
                   'имени': 'поселок имени'}

# получим доступ к элементам словаря, заменим ключ на значение словаря, удалим непечатные символы
for key, value in text_to_replace.items():
    data['locality_name'] = data['locality_name'].str.replace(key, value).str.strip()
```

Так мы отредактировали столбец "название населённого пункта", теперь в нем только названия.

Теперь приведем данные к нужным типам для проведения анализа.

### Изменение типов данных

```python
# зададим цикл для изменения типов данных в определенных столбцах
for i in ['floors_total', 'balcony', 'parks_around3000', 'ponds_around3000', 'days_exposition']:
    data[i] = data[i].astype(int)
data.info()
```

Вещественный тип данных заменен на целочисленный в следующих столбцах:
"всего этажей в доме", "число балконов", "число парков в радиусе 3 км", "число водоёмов в радиусе 3 км", "сколько дней было размещено объявление". В связи с тем, что данные показатели не могут быть дробными.

## Расчёты и добавление результатов в таблицу

### Цена квадратного метра

```python
# создадим столбец "цена квадратного метра",
# для этого значение столбца "цена на момент снятия с публикации" разделим на "площадь квартиры в квадратных метрах"
data['price_square'] = data['last_price'] / data['total_area']

# округлим до двух знаков после запятой
data['price_square'] = data['price_square'].round(2)
```

### День недели, месяц, год публикации

```python
# добавим столбцы с днем недели, месяцем и годом публикации обьявления при помощи метода .dt
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].astype('datetime64[M]')
data['year_exposition'] = data['first_day_exposition'].dt.year
```

### Категория этажа

```python
# пропишем функцию, которая разделит список объявлений на 3 категории: "первый", "последний", "другой".
def categorize_floor(row):
    '''
    функция принимает значения конкретных
    столбцов датафрейма и категоризирует их:
    если floor == 1, то 'первый',
    если floor == floors_total, то 'последний',
    если не соответствует ни одному условию, то 'другой'
    '''
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
    
# при помощи метода apply() применим функцию к датафрейму
data['categorize_floor'] = data.apply(categorize_floor, axis=1)
```

### Отношение жилой площади и кухни к общей

```python
# добавим столбцы с отношением жилой площади к общей, а также площади кухни к общей:
data['living_total_area'] = data['living_area'] / data['total_area']
data['kitchen_total_area'] = data['kitchen_area'] / data['total_area']
```

Посчитали и добавили в таблицу:

- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.

### Изменение цены по степени удалённости от центра

```python
# создадим столбец с расстоянием до центра в километрах: округлим до целых значений
data['center_km'] = data['city_centers_nearest'] // 1000 # // - целочисленное деление
report = (data.query('locality_name == "Санкт-Петербург"').groupby('center_km')
          .agg({'price_square': 'mean'})
          .rename(columns = {'price_square': 'Средняя цена м2'})
         )

report['Средняя цена м2'].plot(grid=True, figsize=(15,3))
plt.axvline(x=7.5, color='red', linestyle = ':') # граница город-область
plt.xticks(report.index.values, report.index.values) # принудительная установка делений на оси х
plt.show()
```

Исходя из полученного графика видно, что цена квадратного метра снижается от самого центра до 3 км, затем начинает увеличиваться до 6 км. Данное явление может быть обусловлено "ржавым поясом" Санкт-Петербурга (это около 130 тыс. га промзон в околоцентральной части Петербурга).
